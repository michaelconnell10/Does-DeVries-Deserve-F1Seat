```python
import fastf1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()

race = fastf1.get_session(2022, 'French Grand Prix', 'FP1')
race.load()

dev = race.laps.pick_driver('DEV')
rus = race.laps.pick_driver('RUS')

fig, ax = plt.subplots()
fig.set_figwidth(15)

ax.plot(dev['LapNumber'], dev['LapTime'], color='red', label='De Vries')
ax.plot(rus['LapNumber'], rus['LapTime'], color='cyan', label='Russell')

ax.set_title("DEV vs RUS")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")

plt.show()
```

```python
### Russell ends up going about half a second quicker than De Vries. Half a second sounds like a lot, especially
### considering Red Bull are looking to find a driver that can consisdently be 3-tenths to Max. But just like the
### Latifi chart, you cannot really get useful data from FP1 - especially when the amount of time in the car is
### massively different between the drivers. I mean, Albon in the Williams (arguably the worst car on the grid) put
### in a faster lap than De Vries during this session:
```

```python
import fastf1
session = fastf1.get_session(2022, 12, 'FP1')

dev_lap = race.laps.pick_driver('DEV').pick_fastest()
rus_lap = race.laps.pick_driver('RUS').pick_fastest()

rus_tel = rus_lap.get_car_data().add_distance()
dev_tel = dev_lap.get_car_data().add_distance()

fig, ax = plt.subplots()
fig.set_figwidth(15)
ax.plot(dev_tel['Distance'], dev_tel['Speed'], color='red', label='DEV')
ax.plot(rus_tel['Distance'], rus_tel['Speed'], color='cyan', label='RUS')

ax.set_xlabel('Distance in m')
ax.set_ylabel('Speed in km/h')

ax.legend()
plt.suptitle(f"Fastest Lap Comparison \n "
             f"{session.event['EventName']} {session.event.year} Free Practice 1")

plt.show()
```

```python
### De Vries loses out most of his time in two main places on the circuit: The complicated section at turns 4-6 and
### the high(ish) speed hairpin of turn 11:
```
